```r
library(tableone)
library("gtsummary")
```

```r
df <- read.csv("cleaned_cgrd_data_no_negatives.csv")
head(df)
```

```r
mydata1=df[which(df$bloc=='1'),]
names(mydata1)
```

```r
mydata=mydata1[c('died_in_hosp',
                 'gender',
                 'age',
                 'Weight_kg',
                 'GCS',
                 'HR',
                 'SysBP',
                 'MeanBP',
                 'DiaBP',
                 'RR',
                 'SpO2',
                 'Temp_C',
                 'FiO2_1',
                 'Potassium',
                 'Sodium',
                 'Chloride',
                 'Glucose',
                 'BUN',
                 'Creatinine',
                 'Magnesium',
                 'Calcium',
                 'Ionised_Ca',
                 'CO2_mEqL',
                 'SGOT',
                 'SGPT',
                 'Total_bili',
                 'Albumin',
                 'Hb',
                 'WBC_count',
                 'Platelets_count',
                 'PTT',
                 'PT',
                 'INR',
                 'Arterial_pH',
                 'paO2',
                 'paCO2',
                 'Arterial_BE',
                 'HCO3',
                 'Arterial_lactate',
                 'mechvent',
                 'Shock_Index',
                 'PaO2_FiO2',
                 'median_dose_vaso',
                 'max_dose_vaso',
                 'input_total',
                 'input_4hourly',
                 'output_total',
                 'output_4hourly',
                 'cumulated_balance',
                 'SOFA',
                 'SIRS')]
```

```r
#Create a variable list which we want in Table 1
listVars <- names(mydata)
listVars

#Define categorical variables
catVars<-listVars[-c(3:51)]
catVars
```

```r
#Total Population
table1 <- CreateTableOne(vars = listVars, data = mydata, factorVars = catVars)
table1
table1 <- print(table1)
```

```r
# Removing Gender from list of variables 
listVar <- listVars
table2 <- CreateTableOne(listVar, mydata, catVars, strata = c("died_in_hosp"))
table2

table2 <- print(table2)
```
